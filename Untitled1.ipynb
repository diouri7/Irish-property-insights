```python
df.columns = [
    "Date",
    "Address",
    "County",
    "Eircode",
    "Price",
    "Not_Full_Market_Price",
    "VAT_Exclusive",
    "Description",
    "Property_Size"
]
```

```python
df["Price"] = (
    df["Price"]
    .astype(str)
    .str.replace(r"[^\d.]", "", regex=True)
    .astype(float)
)
```

```python
df["Price"].head()
df["Price"].dtype
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df["Year"] = df["Date"].dt.year
```

```python
df[["Date", "Year"]].head()
```

```python
county_year = (
    df.groupby(["County", "Year"])["Price"]
    .median()
    .reset_index()
)

county_year.head()
```

```python
price_pivot = county_year.pivot(
    index="County",
    columns="Year",
    values="Price"
)

price_pivot.head()
```

```python
sorted(df["Year"].unique())
```

```python
start_year = 2020
end_year = 2024
n_years = end_year - start_year

price_pivot["CAGR_5Y"] = (
    (price_pivot[end_year] / price_pivot[start_year]) ** (1 / n_years) - 1
)
```

```python
price_pivot[["CAGR_5Y"]].head()
```

```python
returns = price_pivot.loc[:, 2020:2024].pct_change(axis=1)

price_pivot["Volatility_5Y"] = returns.std(axis=1)
```

```python
price_pivot[["CAGR_5Y", "Volatility_5Y"]].head()
```

```python
price_pivot["Score"] = (
    price_pivot["CAGR_5Y"] / price_pivot["Volatility_5Y"]
)
```

```python
final_table = price_pivot[["CAGR_5Y", "Volatility_5Y", "Score"]].sort_values(
    by="Score",
    ascending=False
)

final_table.head(10)
```

```python
volume = (
    df[df["Year"].between(2020, 2024)]
    .groupby("County")
    .size()
    .sort_values(ascending=False)
)

volume.head(10)
```

```python
price_pivot["Liquidity"] = volume
price_pivot["Adjusted_Score"] = (
    price_pivot["Score"] * (price_pivot["Liquidity"] / price_pivot["Liquidity"].max())
)
```

```python
price_pivot = price_pivot.select_dtypes(include="number")
```

```python
# Keep only numeric year columns
year_cols = [col for col in price_pivot.columns if isinstance(col, int)]

price_only = price_pivot[year_cols]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))

for county in ["Dublin", "Laois", "Cork"]:
    plt.plot(
        year_cols,
        price_only.loc[county],
        linewidth=2,
        label=county
    )

plt.title("Irish Property Price Evolution by County", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Median Price (€)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
